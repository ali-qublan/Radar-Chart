```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories = ['Jumping', 'Shooting', 'Drebbling', 'Satam', 'Accuracy']

player_1 = [9, 9, 10, 9, 7]
Player_2 = [10, 10, 9, 10, 9]
Player_3 = [7, 9, 10, 7, 10]
```

```python
# a list that represents the label location in radians
label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(player_1))
```

```python
label_loc
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(polar=True)   # must be used to make a radar chart
plt.plot(label_loc, player_1, label='Player 1')
plt.plot(label_loc, Player_2, label='Player 2')
plt.plot(label_loc, Player_3, label='Player 3')
plt.title('Restaurant comparison', size=20)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories) #is used to place the category names on label locations
plt.legend()
plt.show()
```
